```python
# condition = df['City'] == 'Chicago'

# condition = user_format.lower() in df['Split']

# FUNCTIONS:

def split_str_list(entry:str, seperator:str):
    split = []
    split2 = []
    f = entry.split(seperator)
    split.append(f)
    for element in split:
        for item in element:
            split2.append(item.lower())

    return split2

def certain_format(df, d_column:str, user_format:str):
    # elements_df = pd.DataFrame(columns=df.columns.to_list())
    for row in df[d_column]:
        df['Split'] = split_str_list(row, ', ')
        condition = user_format.lower() in df['Split']
    return df[condition]
```
